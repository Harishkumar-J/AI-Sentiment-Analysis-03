```python
import pandas as pd
import regex as re
import numpy as np

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

import transformers
from transformers import pipeline
```

```python
data = pd.read_csv("book_reviews_sample.csv")
data.head()
```

```python
data['reviewText'][0]
```

```python
data['reviewText_clean'] = data['reviewText'].str.lower()
```

```python
data['reviewText_clean'] = data['reviewText_clean'].apply(lambda x: re.sub(r'[^\w\s]', '', str(x)))
```

```python
data.head()
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
data['vader_sentiment_score'] = data['reviewText_clean'].apply(lambda review : vader_sentiment.polarity_scores(review)['compound'])
```

```python
data.head()
```

## Bins: the value ranges that define how the scores are divided

```python
bins = [-1,-0.1,0.1,1]
names = ['negative','neutral','positive']

data ['vader_sentiment_label'] = pd.cut(data['vader_sentiment_score'],bins, labels=names)
```
